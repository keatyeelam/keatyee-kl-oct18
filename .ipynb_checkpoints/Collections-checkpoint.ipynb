##sample 2

```python
names = ['csev','cwen','zqian','cwen']
```

```python
counts = dict()
for name in names:
    if name not in counts:
        counts[name] = 1
    else :
        counts[name] = counts[name] + 1
```

```python
print(counts)
```

```python
counts = dict()
for name in names:
    counts[name] = counts.get(name,0)+1
print(counts)
```
